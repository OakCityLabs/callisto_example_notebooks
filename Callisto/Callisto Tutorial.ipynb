# Welcome to Callisto!

This notebook is meant to give you a brief tour of the many features that Callisto offers to make data science easy.

## Recent Notebooks
On the left side of the screen, you should see a 'Recent Notebooks' section. As the title suggests, this will provide a list of your most recently active notebooks, whether they are in the cloud or on your machine.

## Tutorials
Further down in the sidebar, you will see a list of tutorials (where you probably found this notebook) that exist to give you a sense of what's possible with Callisto. If you are a beginner with Jupyter Notebooks and data science, you may find these useful as references.

## Servers
In the top right of the screen, you will see a "Servers" button. 

Click this, and you will see a screen similar to the one shown below. This will provide information about your currently running servers. If you have remote servers running in the cloud, or custom servers, they will show up here as well.
<br><br>
When a server is selected, the middle of the screen shows the working directory for that server. You can open notebooks by navigating to them in this file directory and double clicking them.
<br><br>
On the right side of the screen, information about the server is shown, like the packages installed and the current status.

## Data Explorer
If you open a notebook and take another look at the top right corner of the screen, you will see a button called "Data Explorer". This opens a new window that shows the current kernel's stored variables.
<br>
You can click the buttons next to the variables to get more information about them, or to copy their values to the clipboard.

## Editing the notebook
Between each cell in a notebook, you will see the following buttons:

- **Insert**: insert a new cell
- **Paste**: paste a cell here
- **Merge**: combine the text in the cells above and below the button
- **Swap**: move the bottom cell up, and the top cell down

You will also see the following buttons on the right side of each cell:

- The top one is used for dragging the cell around. You can click and hold a cell here, and move it above or below other cells easily.
- The trash can in the middle is used to delete the cell.
- The three dots on the bottom give you more options, including the ability to **cut**, **copy**, **duplicate**, **change to Markdown/Code**, and **split at cursor**

And last, but not least, you will also see the **share** button next to each cell:

This gives you the ability to send the code/Markdown to other applications, including mail, notes, messages, etc. You can also copy to the clipboard, which is very useful for sharing plots and images generated by your analysis.
