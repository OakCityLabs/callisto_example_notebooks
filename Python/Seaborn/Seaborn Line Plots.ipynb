# Statistical Relationships With Seaborn - Line Plots
### Use the **relplot()** and **catplot()** functions from the **Seaborn** library to visualize the relationship between your variables.
### See more info at [https://seaborn.pydata.org/tutorial/relational.html](https://seaborn.pydata.org/tutorial/relational.html)

```python
# Import the Seaborn Library
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
# Download data from Seaborn
# If you get an SSL error when you try to load the dataset, uncomment the following 2 lines:
# import ssl
# ssl._create_default_https_context = ssl._create_unverified_context

fmri = sns.load_dataset("fmri")
```

# Line Plot for Numerical Data

```python
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri);
```

# Line Plot for Numerical Data with Facets
### If you need to visualize how a third variable interacts with your data, you can plot subsets of the data on multiple axes:

```python
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            col="subject", col_wrap=5,
            height=3, aspect=.75, linewidth=2.5,
            kind="line", data=fmri.query("region == 'frontal'"));
```
