# Statistical Relationships With Seaborn - Scatter Plots
### Use the **relplot()** and **catplot()** functions from the **Seaborn** library to visualize the relationship between your variables.
### See more info at [https://seaborn.pydata.org/tutorial/relational.html](https://seaborn.pydata.org/tutorial/relational.html)

```python
# Import the Seaborn Library
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
# Download data from Seaborn
# If you get an SSL error when you try to load the dataset, uncomment the following 2 lines:
# import ssl
# ssl._create_default_https_context = ssl._create_unverified_context

tips = sns.load_dataset("tips")
```

# Scatter Plot for Numerical Data

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="size")
```

# Scatter Plot for Categorical Data

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips)
```
